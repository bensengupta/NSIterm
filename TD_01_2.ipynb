## Exercice I : Implémenter les structures

1. Proposer ci dessous une implémentation en python d'une liste chaînée (fonctions **initL**, **cons**, **addL**, **popL**, **isEmptyL**, **lenL**).

```python
# question 1
def initL():
    return None

def cons(element, L):
    return [element, L]

def addL(element, L):
    return cons(element,L)

def popL(L):
    """
    supprime l'élément en tête de liste et le renvoie
    renvoie également la nouvelle liste
    """
    assert L is not None
    return L[0],L[1]

def isEmptyL(L):
    return L is None

def lenL(L):
    """
    renvoie le nombre d'éléments d'une liste
    """
    i = 0
    temp = L
    while temp is not None:
        i+=1
        temp = temp[1]
    return i
```

2. Utiliser la structure de liste chaînée pour construire une structure de pile (fonctions **initS**, **pushS**, **popS**, **topS**, **isEmptyS**).

```python
# question 2
def initS():
    return initL()

def pushS(element,s):
    return addL(element,s)

def popS(s):
    return popL(s)

def topS(s):
    return s[0]

def isEmptyS(s):
    return isEmptyL(s)
```

3. Utiliser 2 structures de pile précédente afin de créer une structure de file (fonctions **initQ**, **enQueue**, **deQueue**, **topQ**, **isEmptyQ**)

```python
# question 3
def initQ():
    return [initS(),initS()]

def enQueue(element, q):
    return [pushS(element,q[0]),q[1]]

def deQueue(q):
    assert not isEmptyQ(q), "impossible, la file est vide"
    if isEmptyS(q[1]):
        while not isEmptyS(q[0]):
            val,q[0] = popS(q[0])
            q[1] = pushS(val,q[1])
    val,q[1] = popS(q[1])
    return val, [q[0],q[1]]

def topQ(q):
    assert not isEmptyQ(q)
    if isEmptyS(q[1]):
        while not isEmptyS(q[0]):
            val,q[0] = popS(q[0])
            q[1] = pushS(val,q[1])
    val = topS(q[1])
    return val # Attention, la file n'est pas modifiée!

def isEmptyQ(q):
    return isEmptyS(q[0]) and isEmptyS(q[1])

        
```
